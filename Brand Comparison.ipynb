=========================================================================================================================

                                             Categorizing Discount Price

=========================================================================================================================

```python
mobile_data['Discount_Price1'].describe()
```

```python
def segments(dis_price1):
    if dis_price1 >1500 :
        return "Highest Discount"
    elif dis_price1 <= 500:
        return "Lowest Discount"
    elif dis_price1 ==1500:
        return "Average Dicount"
    else:
        return " Moderate/No Discount"
```

```python
mobile_data['Segments1']=mobile_data['Discount_Price1'].apply(segments)
```

```python
size=mobile_data['Segments1'].value_counts()
labels=np.unique(mobile_data.Segments1)
plt.figure(figsize=(6,6))
plt.title("Realme Mobile \n Categorizing Discount Price  \n:")
plt.pie(size, labels = labels,autopct='%1.1f%%')
```
